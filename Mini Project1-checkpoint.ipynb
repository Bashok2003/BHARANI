```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\91799\Downloads\vgsales.csv.zip")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
print("Na values:")
df.isna().sum()
```

```python
df.dropna(subset = ['Year'], inplace=True)
```

```python
print("Number of duplicates:")
df.duplicated().sum()
```

```python
df_outliers = df[['Name', 'Year']].sort_values(by='Year')
df_outliers
```

```python
df.drop(df[df.Year > 2018].index, inplace=True)
```

```python
sns.set_theme(style="white", palette="pastel")

df_sales = df[['Year', 'Global_Sales', 'Genre']].groupby(['Year', 'Genre']).sum()

ax = sns.lineplot(data=df_sales, x='Year', y= 'Global_Sales', hue = 'Genre').set(title='Global Sales by Genre')
plt.figure(figsize=(6,4))
```

```python
df_games = df[['Name', 'Global_Sales', 'Publisher']].sort_values('Global_Sales',ascending=False).head(10)
```

```python
sns.set_theme(style="white", palette="pastel")
sns.barplot(data=df_games, x='Global_Sales', y='Name', orient='h').set(title='Top 10 Games by Global Sales')
plt.xlabel("Global Sales")
```

```python
genre_counts = df['Genre'].value_counts()
plt.figure(figsize = (15,8))
plt.pie(genre_counts.values, labels = genre_counts.index, autopct='%1.1f%%')
plt.title('Frequency by Genre')
plt.show()
```

```python
genres1=df[["Genre","NA_Sales","EU_Sales","JP_Sales","Global_Sales"]].pivot_table(values=["NA_Sales","EU_Sales","JP_Sales","Global_Sales"],
    index=None,
    columns=["Genre"],
    aggfunc='sum')
```

```python
genres1.T.plot.bar()
plt.title("Genre Sales by Region")
plt.xlabel("Genre")
plt.ylabel("Sales in Millions")
```
